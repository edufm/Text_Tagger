```python
import pandas as pd
import numpy as np
```

```python
name = "Tweets_USA.csv"
dataset = pd.read_csv(f"./datasets/{name}")
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset = dataset[dataset['Tweet language (ISO 639-1)'] == "en"]
```

```python
to_drop = ['Tweet Id', 'Date', 'Hour', 'User Name', 'Nickname', 'Bio', 'Favs', 'RTs', 'Country',
           'Place (as appears on Bio)', 'Profile picture', 'Followers',
           'Following', 'Listed', 'Tweet language (ISO 639-1)', 'Tweet Url']
dataset = dataset.drop(to_drop, axis=1)
dataset.head()
```

```python
dataset.to_csv(f"./datasets_samples/{name}")
dataset.loc[np.random.permutation(np.array(dataset.index))[:5000]].to_csv(f"./datasets_samples/{name}")
```
